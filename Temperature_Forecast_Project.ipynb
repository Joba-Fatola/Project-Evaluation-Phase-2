```python
# for model buidling

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.svm import LinearSVR,SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.model_selection import train_test_split,cross_val_score,cross_validate,ShuffleSplit,GridSearchCV,KFold,RandomizedSearchCV
from sklearn.metrics import r2_score,mean_squared_error

import pickle
```

```python
# Split the data into train and test

# Statify the splitted data to ensure equal distribution of each feature for both Train and Validation
train_set, test_set = train_test_split(df_temp, test_size=0.2, random_state=42)
print(train_set.shape, test_set.shape)
```

```python
# Split the data into train and test

x_train=train_set.drop(columns=['Next_Tmax','Next_Tmin'])
y_Tmin_train=train_set['Next_Tmin']
y_Tmax_train=train_set['Next_Tmax']
x_test=test_set.drop(columns=['Next_Tmax','Next_Tmin'])
y_Tmin_test=test_set['Next_Tmin']
y_Tmax_test=test_set['Next_Tmax']

# Scale the feature sets
scaler = StandardScaler()
x_train = pd.DataFrame(scaler.fit_transform(x_train))
x_test = pd.DataFrame(scaler.transform(x_test))
```

## Model Training, Evaluation and Comparison

```python
Machine_learning_algorithm = [LinearRegression(),LinearSVR(max_iter=5000), SVR(max_iter=2000), DecisionTreeRegressor(),
                              RandomForestRegressor(),MLPRegressor(max_iter=2000),
                              KNeighborsRegressor(),GradientBoostingRegressor()]
mean_squared_error_min = []
r2_score_min = []
mean_squared_error_max = []
r2_score_max = []
mean_squared_error_min_ofc = []
r2_score_min_ofc = []
mean_squared_error_max_ofc = []
r2_score_max_ofc = []


for models in Machine_learning_algorithm:
    classifier_min = models.fit(x_train, y_Tmin_train)
    y_pred_min = classifier_min.predict(x_test)
    y_pred_min_ofc = classifier_min.predict(x_train)
    classifier_max = models.fit(x_train, y_Tmax_train)
    y_pred_max = classifier_max.predict(x_test)
    y_pred_max_ofc = classifier_max.predict(x_train)
    
    mean_squared_error_min.append(mean_squared_error(y_Tmin_test, y_pred_min))
    r2_score_min.append(r2_score(y_Tmin_test, y_pred_min))
    mean_squared_error_max.append(mean_squared_error(y_Tmax_test, y_pred_max))
    r2_score_max.append(r2_score(y_Tmax_test, y_pred_max))
    mean_squared_error_min_ofc.append(mean_squared_error(y_Tmin_train, y_pred_min_ofc))
    r2_score_min_ofc.append(r2_score(y_Tmin_train, y_pred_min_ofc))
    mean_squared_error_max_ofc.append(mean_squared_error(y_Tmax_train, y_pred_max_ofc))
    r2_score_max_ofc.append(r2_score(y_Tmax_train, y_pred_max_ofc))
    
classification_report = pd.DataFrame({
    "ML Algorithm": ["Linear Regression","Linear SVM", "SVM", "DecisionTree", 
                     "RandomForest","MLP Regression","KNeighbors  Regression","GradientBoosting Regression"],
    "Root Mean Square error Temp min (Lower is better)": mean_squared_error_min,
    "Root Mean Square error Temp min (Overfitting check)": mean_squared_error_min_ofc,
    "Root Mean Square error Temp max (Lower is better)": mean_squared_error_max,
    "Root Mean Square error Temp max (Overfitting check)": mean_squared_error_max_ofc,
    "R square Temp min (Higher is better)": r2_score_min,
    "R square Temp min (Overfitting check)": r2_score_min_ofc,
    "R square Temp max (Higher is better)": r2_score_max,
    "R square Temp max (Overfitting check)": r2_score_max_ofc
})

classification_report.set_index('ML Algorithm', inplace=True)
classification_report
```

**Root Mean Square Error (RMSE):**

* Indicates how far off the predictions are from the actual values.
* Lower values are preferred, as they suggest smaller prediction errors.

**RMSE (Overfitting Check):**

* These values represent RMSE calculated on the training data.
* Used as a check for overfitting; extremely low values could indicate overfitting.

**R-Square (R²):**

* Measures how well the model explains the variance in the data.
* Higher values are better, with 1 indicating a perfect fit and 0 indicating a poor fit.

Here's a summary of the algorithm performance:

**1. Linear Regression:**
* RMSE values are moderate, indicating some prediction error.
* R² values are decent, suggesting a reasonable fit.
* Overfitting check RMSE is slightly lower than test RMSE.

**2. Linear SVM (Support Vector Machine):**
* Similar performance to Linear Regression.
* Moderate RMSE and decent R² values.
* Overfitting check RMSE is slightly lower than test RMSE.

**3. SVM (Support Vector Machine):**
* Significantly lower RMSE compared to Linear models, indicating better prediction accuracy.
* High R² values suggest a good fit.
* Overfitting check RMSE is also low, indicating good generalization.

**4. Decision Tree:**
* Extremely low RMSE on training data indicates overfitting.
* High RMSE on test data suggests poor generalization.
* R² on training data is 1, a clear sign of overfitting.

**5. Random Forest:**
* Low RMSE on both training and test data.
* High R² values, indicating a good fit.
* Suggests that Random Forest performs well without overfitting.

**6. MLP Regression (Multi-Layer Perceptron):**
* Low RMSE on both training and test data.
* High R² values, indicating a good fit.
* Indicates that the neural network model performs well without overfitting.

**7. K-Neighbors Regression:**
* Moderate RMSE values.

**8. Gradient Boosting Regression:**

In summary, SVM appears to have the best performance in terms of RMSE and R², indicating higher prediction accuracy and a better fit.
